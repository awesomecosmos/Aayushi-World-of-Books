```python
import requests
import json
import numpy as np
import re
from utils import *
```

```python
my_request = 'https://openlibrary.org/search.json?title=geek+girl&author=smale+holly'
response = requests.get(my_request)
json_result = response.json()
lst_to_filter = json_result['docs'][0]['seed']
filtered_books = list(filter(lambda v: re.match('/books/*', v), lst_to_filter))
```

```python
for book in filtered_books:
    my_request = f'https://openlibrary.org{book}.json'
    response = requests.get(my_request)
    json_result = response.json()
    for key, value in json_result.items():
        print(key, ':', value)
```
